```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data_df = pd.DataFrame(purchase_data)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
# Calculate the total number of players 
unique_players = len(purchase_data_df["SN"].unique())
unique_players



```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Run calculations to obtain number of unique items, number of purchases, average price, total revenue
unique_items = len(purchase_data_df["Item ID"].unique())
unique_items

total_purchase = purchase_data_df["Purchase ID"].count()
total_purchase

total_revenue = purchase_data_df["Price"].sum()
total_revenue 

average_pricep = purchase_data_df["Price"].mean()
average_pricep 

average_pricer = total_revenue/total_purchase
average_pricer



# Display a summary data frame to hold the results
purchase_analysis_df = pd.DataFrame ([{"Number of Unique Items": unique_items, "Average Price": average_pricep, "Number of Purchase": total_purchase,
                                      "Total Revenue": total_revenue}])


# Formating "Average Price" and "Total revenue"
purchase_analysis_df["Average Price"] = purchase_analysis_df["Average Price"].map("${:,.2f}".format)
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map("${:,.2f}".format)
purchase_analysis_df


# Change the order of the columns
org_purchase_analysis_df = purchase_analysis_df [["Number of Unique Items","Number of Purchase", "Average Price", "Total Revenue"]]
org_purchase_analysis_df 

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Drop the duplicates 
gender_demo_df = purchase_data [["Gender", "SN"]].drop_duplicates()
gender_demo_df 

gender_demo_new = gender_demo_df.rename(columns={'SN': 'Total Count', 'Gender': 'Gender'})

# Calculate percentage per type of gender 

P = gender_demo_new.groupby('Gender')['Total Count'].count().reset_index()
P ['Percentage'] = 100 * P['Total Count'] / P['Total Count'].sum()

# Rouding percentage 
P['Percentage'] = P['Percentage'].round(2)
P

# Formating the percentage column to add the % sign
P['Percentage'] = P["Percentage"].map("{:,.2f}%".format)
P
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Separate the data into fields according to "Gender" values using GroupBy 
grouped_gender_purchased_data_df = purchase_data_df.groupby(["Gender"])

# print(grouped_gender_purchased_data_df)
# Using a data function to be able to visualize the the "groupby" object. 
grouped_gender_purchased_data_df["Purchase ID"].count().head()



# Get total purchase price by gender 
total_purchase_price = grouped_gender_purchased_data_df["Price"].sum()
total_purchase_price.head()
cln_total_purchase_price = total_purchase_price.map ("${:,.2f}".format)
cln_total_purchase_price.head()


# Calculates the average purchase price by gender
avg_purchase_price = grouped_gender_purchased_data_df["Price"].mean()
avg_purchase_price.head()
cln_avg_purchase_price = avg_purchase_price.map("${:,.2f}".format)
cln_avg_purchase_price.head()


# Normalize the calculated totals (total purchase price divided by purchase count by gender)
normalized_totals = total_purchase_price/grouped_gender_purchased_data_df["Purchase ID"].count()
cln_normalized_totals = normalized_totals.map("{:,.2f}".format)
cln_normalized_totals.head()


# Display the summary for gender data, get all columns to display Data Frame and add needed columns to it
org_gender_purchased_data_df = pd.DataFrame(grouped_gender_purchased_data_df["Purchase ID"].count())
org_gender_purchased_data_df["Average Purchase Price"] = cln_avg_purchase_price  
org_gender_purchased_data_df["Total Purchase Value"] = cln_total_purchase_price 
org_gender_purchased_data_df["Normalized Totals"] = cln_normalized_totals 
org_gender_purchased_data_df


# DataFrame analysis, summary of purchases grouped by gender, rename "Purchase ID" column to Purchase Count 
summary_gender_purchased_data_df = org_gender_purchased_data_df.rename(columns={"Purchase ID":"Purchase Count"})
summary_gender_purchased_data_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Using age bins to categorize the existing players
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_bins,labels=group_names)

# Group by age and calculate, calculate the number and percentages (rounds the percentage for two decimal points)
age_demog_df = purchase_data.groupby("Age Group")
age_demog_count = age_demog_df["SN"].nunique()
age_demog_perc = round((age_demog_count/unique_players)*100,2)

# Display Age Demographics Table
age_demog_summary = pd.DataFrame({"Percentage of Players":age_demog_perc,"Total Count":age_demog_count})
age_demog_summary
```

## Top Spenders

* Display a preview of the summary data frame



```python
# Create the Data Frame and group by players/spendors (SN)
sn_purchase_data_df = pd.DataFrame(purchase_data)
sn_purchase_data_df 

# Group by spendors (SN)
group_SN_top_spender_df = sn_purchase_data_df.groupby("SN")
group_SN_top_spender_df.count()

# In the Data Frame grouped by "SN" get how many times each player has purchased a game 
amount_by_spendor_df = pd.DataFrame(group_SN_top_spender_df["Purchase ID"].count())
amount_by_spendor_df 

# Calculate total purchase price by "SN" 
total_purchase_price_SN = group_SN_top_spender_df["Price"].sum() 
total_purchase_price_SN

# Calculate average purchase price by "SN"
avg_purchase_price_SN = group_SN_top_spender_df["Price"].mean()
avg_purchase_price_SN  

# Format average purchase price to display the "$" sign and 2 decimal points 
cln_avg_purchase_price_SN = avg_purchase_price_SN.map("${:,.2f}".format) 
cln_avg_purchase_price_SN 


# Create a summary Data Frame top spenders grouped by "SN" to hold the results and add columns if needed 
amount_by_spendor_df["Average Purchase Price"] = cln_avg_purchase_price_SN 
amount_by_spendor_df["Total Purchase Price"] = total_purchase_price_SN
amount_by_spendor_df 

# Order the summary by total purchase price in descending order and rename columns 
desc_purchased_data_df = amount_by_spendor_df.rename(columns={"Purchase ID": "Purchase Count", "Total Purchase Price" : "Total Purchase Value"})
preview_spendor_df = desc_purchased_data_df.sort_values("Total Purchase Value", ascending=False)
preview_spendor_df.head() 

# Format the total purchase value adding the "$" sign and rename columns
cln_total_purchase_price_SN = total_purchase_price_SN.map("${:,.2f}".format) 

desc_purchased_data_df = amount_by_spendor_df.rename(columns={"Purchase ID": "Purchase Count"})
preview_spendor_df["Total Purchase Value"] = cln_total_purchase_price_SN
preview_spendor_df.head() 

```
